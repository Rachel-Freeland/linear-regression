### Load data set into dataframe
### Show the first 5 rows

```python
df = pd.read_csv('/kaggle/input/cricket-chirp-vs-temperature/Cricket_chirps.csv')
df.head()
```

### Show the `description` and the `info` of the data set

```python
df.describe()
```

```python
df.info()
```

```python
X = df.X.values
X_round = X.round(1)
X_round
```

```python
y = df.Y.values
y
```

### Using a regression model, split your data into train and test portions

```python
plt.scatter(X_round, y)
```

```python
temp = X_round.reshape(-1, 1)
temp
```

### Setup the model

```python
model = LinearRegression().fit(temp, y)
model
```

```python
predictions = model.predict(temp)
rounded_predictions = predictions.round(1)
rounded_predictions
```

```python
plt.scatter(X_round, y)
plt.plot(X_round, predictions, color="red")
```

```python
temp_predict = 99
chirps_predict = model.predict([ [temp_predict] ])[0]
print(f'When the temperature increases to {temp_predict} the predicted chirps per second will be {round((chirps_predict), 1)} chirps.')
```

### Train the Model

```python
x_train, x_test, y_train, y_test = train_test_split(temp, y, train_size=.8, test_size=.2)
x_train
```

```python
print(f'x_train shape {x_train.shape}')
print(f'y_train shape {y_train.shape}')
print(f'x_test shape {x_test.shape}')
print(f'y_test shape {y_test.shape}')
```

### Plot Training Data

```python
plt.scatter(x_train, y_train, color="purple")
plt.xlabel('Temp in F')
plt.ylabel('Chirps/sec')
plt.title('Training Data')
```

### Plot Testing Data

```python
plt.scatter(x_test, y_test, color="green")
plt.xlabel('Temp in F')
plt.ylabel('Chirps/sec')
plt.title('Testing Data')
```

```python
lm = LinearRegression()
lm.fit(x_train, y_train)
y_prediction = lm.predict(x_test)
print(f'Training accuracy {round(lm.score(x_train, y_train) * 100, 2)}%')
print(f'Testing accuracy {round(lm.score(x_test, y_test) * 100, 2)}%')
```

```python
plt.scatter(x_train, y_train, color="magenta")
plt.plot(x_test, y_prediction)
plt.xlabel("Temp in F")
plt.ylabel("Chirps/sec")
plt.title("Trained model plot")
plt.plot
```

### Conclusions:
1. This model does not show a very good correlation between temperature and the chirps per second for crickets
2. This model and its data roughly achieves a 50% accurracy for predicting chirps vs temp
